<h3>Decision Tree Algorithm on the Iris dataset</h3>

```python
import sklearn.datasets as datasets
import pandas as pd

# Loading the iris dataset
dataset=datasets.load_iris()

# Forming the iris dataframe
df=pd.DataFrame(dataset.data, columns=dataset.feature_names)
print(df.head(5))

y=dataset.target
print(y)
```

<h3>Now let us define the Decision Tree Algorithm</h3>

```python
from sklearn.tree import DecisionTreeClassifier
dtree=DecisionTreeClassifier()
dtree.fit(df,y)

print('Decision Tree Classifer Created')
```

<h3>Train Test Split</h3>

```python
from sklearn.model_selection import train_test_split  #to split the dataset for training and testing
iris = pd.read_csv('Iris.csv')
```

```python
X = iris.drop(['Id', 'Species'], axis=1)
y = iris['Species']
# print(X.head())
print(X.shape)
# print(y.head())
print(y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=10)
print(X_train.shape)
print(y_train.shape)
print(X_test.shape)
print(y_test.shape)
```

<h3>Time to check the accuracy of the model</h3>

```python
from sklearn import metrics
dt = DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred = dt.predict(X_test)
acc_dt = metrics.accuracy_score(y_pred,y_test)
print('The accuracy of the Decision Tree is', acc_dt)
```
